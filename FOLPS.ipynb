```python
from desilike.observables.galaxy_clustering import TracerPowerSpectrumMultipolesObservable
theory = FOLPSTracerPowerSpectrumMultipoles(template=template)
observable = TracerPowerSpectrumMultipolesObservable(
             data=pk,
             covariance=covariance_data,
             klim={0: [0.00379811, 0.3175159, 0.0049], 2: [0.00379811, 0.3175159, 0.0049], 4:[0.00379811, 0.3175159, 0.0049]},  
             theory=theory)
```

```python
from desilike.likelihoods import ObservablesGaussianLikelihood

likelihood = ObservablesGaussianLikelihood(observables = [observable], covariance = covariance_data)
```

```python
print(likelihood.varied_params)
#print(likelihood.flattheory)
```

```python
print(likelihood.flatdata.shape)
```

```python
from desilike import setup_logging
setup_logging()
```

```python
likelihood.runtime_info.pipeline.calculators
```

```python
likelihood()
```

```python
from desilike.emulators import Emulator, EmulatedCalculator, TaylorEmulatorEngine

emulator = Emulator(theory, engine=TaylorEmulatorEngine(order={'*': 2, 'sn0': 1}))  # order 2 except for sn0 (order 1 is enough)
emulator.set_samples()
emulator.fit()
emulator.plot(name='power')
```

```python
from desilike.samplers import EmceeSampler, ZeusSampler
```

```python
from desilike.samplers import ZeusSampler

observable.init.update(theory=emulator.to_calculator())

sampler = ZeusSampler(likelihood, save_fn = 'test_QSO_z1.45', seed=42)
sampler.run(check={'max_eigen_gr': 0.15})
```
